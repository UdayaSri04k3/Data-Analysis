```python
#Splitting arrays
A=np.arange(18).reshape(6,3)#Based on rows and columns the arrays will split into two equal parts.
print(A)
print("After splitting")
#It is to cut as we want like as we want to cut anywhere.
print(np.vsplit(A,(2,5)))#Vertical split
print(np.vsplit(A,(2,4)))# ''
print(np.vsplit(A,(2,3)))# ''
print(np.hsplit(A,(2,5)))#Horizontal split
```
