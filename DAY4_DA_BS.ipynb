```python
d = "https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
```

```python
pag=requests.get(d)
```

```python
sou=BeautifulSoup(pag.text,'html')
print(sou)
```

```python
tabl=sou.find('table')
print(tabl)
```

```python
h=tabl.find_all('th')
print(h)
```

```python
heading=[i.text for i in h]
print(heading)
```

```python
import pandas as pd
```

```python
mydf=pd.DataFrame(columns=heading)
print(mydf)
```

```python
rows=tabl.find_all('tr')
print(rows)
```

```python
data = tabl.find_all('td')
print(data)
```

```python
data_act = [i.text for i in data]
print(data_act)
```

```python
data_act[0:4]
```

```python
df1=pd.DataFrames(columns=heading)
```

```python
st=0
sp=4
for i in range(len(data_act)//4):
  mydf.loc[i]=data_act[st:sp]
  st=st+4
  sp=sp+4
mydf
```

```python
t='https://en.wikipedia.org/wiki/List_of_companies_of_India'
p=requests.get(t)
```

```python
table=c.find('table')
print(table)
```

```python
h=table.find_all('th')
print(h)
```

```python
hd=[i.text for i in h]
print(hd)
```

```python
import pandas as pd
df1=pd.DataFrame(columns=hd)
print(df1)
```

```python
rows=table.find_all('tr')
print(rows)
```

```python
da=table.find_all(t)
print(da)
```

```python
dat_act=[i.text for i in da]
print(dat_act)
```

```python
st=0
sp=0
for i in range(len(dat_act)//9):
  df1.loc[i]=dat_act[st:sp]
  st=st+9
  sp=sp+9
df1
```
