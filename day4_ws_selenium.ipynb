```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
#Define option and set browser capabilities
options = webdriver.ChromeOptions()
#options.add_argument(---some-option)

#Create WebDriver instance with options
driver = webdriver.Chrome(options=options)
```

```python
# Navigate to a website
driver.get("https://www.myntra.com/")
```

```python
search=driver.find_element(By.XPATH,".//input[@class='desktop-searchBar']")
```

```python
search.send_keys("Smart watches")
```

```python
search.send_keys(Keys.ENTER)
```

```python
search.send_keys("shoe")
```

```python
search.send_keys("Girls T-shirt")
```

```python
brand=driver.find_elements(By.XPATH,".//h3[@class='product-brand']")
brands=[i.text for i in brand]
```

```python
print(brands)
print(len(brands))
```

```python
desc=driver.find_elements(By.XPATH,".//h4[@class='product-product']")
descs=[i.text for i in desc]
```

```python
print(descs)
print(len(descs))
```

```python
price=driver.find_elements(By.XPATH,".//span[@class='product-discountedPrice']")
prices=[i.text for i in price]
print(prices)
```

```python
print(prices)
print(len(prices))
```

```python
df=pd.DataFrame(columns=["Brands","Desc","Prices"])
df
```

```python
df['Brands']=brands
df['Desc']=descs
df['Prices']=prices
df
```
