```
#Take a salary from the user if the salary is more then 50000 there is a ten percent tax print his final salary
sal=int(input("Enter salary"))
if sal>=50000:
  f=(sal*10)/100;
  s=sal-f;
  print("The final salary is",s)
else:
  print("The salary is",sal)
```

```
#LC
#newlist=[(st of if) if (code) else (st of else)]
l1=[23,67,12,34,69]
res2=[i*i if i>50 else 'X' for i in l1]
print(res2)
```
