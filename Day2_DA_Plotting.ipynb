```python
#There are five subjects psychology,nano technology,medicine,oncology,anthropology two students have applied to do masters in any of
#these five fields with a line plot show in which field they can excel . The cutoff for each subject is 70
student1=[67,45,73,71,57]
student2=[43,67,68,53,85]
sub=["Psychology","Nano technology","Medicine","Oncology","Anthropology"]
cutoff=[70,70,70,70,70]
#student1
#plt.plottype(X,Y,label,color,marker,linewisth,linestyle)
plt.scatter(sub,student1,label="student1",color="green",marker="p")
#student2
plt.scatter(sub,student2,label="student2",color="red",marker="H")
#cutoff
plt.plot(sub,cutoff,label="cutoff",linestyle=":",linewidth=3) #linestyle is not applicable for scatter
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
```

```python
#An India and Australia match is on live. India has given 298 as target for Australia ,the runs for every 5 over is given as follows
#India=[25,51,84,131,160,189,220,250,257,297]
#Australia have completed 25 overs
#Australia=[15,41,94,110,151]
#plot overs vs runs for both the countries with India as a blue line and Australia as yellow line with required labels and legend.The
# marker for India is Flag and Australia is plus. Take overs from arange function create runs as arrays.
#marker=">"==flag
import numpy as np
oo=np.arange(5,51,5)
rr=np.arange(5,26,5)
india=np.array([25,51,84,131,160,189,220,250,267,297])
australia=np.array([15,41,94,110,151])
plt.title("India vs Australia")
plt.plot( oo,india,label="india",color='b',marker='<')
plt.plot( rr,austraila,label="australia", color='y',marker="p")#line grapy
plt.xlabel('overs')
plt.ylabel('runs')
plt.legend()


```
